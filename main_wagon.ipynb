```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
from tqdm import tqdm

pd.set_option('display.max_colwidth', None)
pd.set_option('display.float_format', '{:.2f}'.format)
pd.options.display.expand_frame_repr = False
```

# парсинг

```python
gr_vagon = {2: 'Крытые вагоны', 4: 'Вагоны-платформы', 6: 'Полувагоны', 7: 'Вагоны-цистерны', 8: 'Изотермические вагоны', 9: 'Прочие вагоны'}
gr_i = 9
```

```python
url = f'https://vagon.by/railcars/list/{gr_i}'
page = requests.get(url)
print(page.status_code)
soup = BeautifulSoup(page.text, "html.parser")
t = soup.find('table')
```

```python
rows = []

for i in tqdm(t.findAll('a')):
    url_vagon = 'https://vagon.by' + i.get('href')
    page_vagon = requests.get(url_vagon)
    vagon_info = dict()

    if page.status_code != 200:
        print(url_vagon, 'page.status_code:', page.status_code)
    soup_vagon = BeautifulSoup(page_vagon.text, "html.parser")
    t_vagon = soup_vagon.findAll('table')
    if len(t_vagon) != 2:
        print(f'wrong number of tables ({len(t_vagon)})', url_vagon, soup_vagon.findAll('h2'))
    vagon_info = pd.read_html(str(t_vagon[0]))[0]
    vagon_info[0] = vagon_info.apply(lambda x: x[0][:-1] if x[0][-1] == ':' else x[0], axis=1)
    vagon_info = vagon_info.set_index(0)
    vagon_info = vagon_info.to_dict()[1]
    vagon_info['url'] = url_vagon

    vagon_info.pop('Внутренние размеры вагона', None)
    vagon_info.pop('Наружные размеры вагона', None)

    if 'Описание на сайте завода-изготовителя' in vagon_info:
        vagon_info['Описание на сайте завода-изготовителя'] = t_vagon[0].find('a', target='_blank').get('href')

    rows.append(vagon_info)
```

```python
df = pd.DataFrame(rows)
df['Группа вагонов'] = gr_vagon[gr_i]
df.to_excel(f'excel_files/database_{gr_vagon[gr_i]}.xlsx', index=False)
df
```
